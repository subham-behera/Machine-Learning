#### To build a linear model given by Y=mX+C

#### Generate noisy data points for y=3x+4 with noise having normal distribution mean=0 and deviation=2

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(0,10,100)
noise=np.random.normal(0,2,100)
plt.plot(noise)
plt.show()
```

```python
Y=3*X+noise
plt.plot(X,Y)
plt.show()
```

#### We need to split the data points into training points

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
```

```python
xtest.shape
```

```python
xtrain.shape
```

#### Build a linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
xtrain.shape
```

```python
#convert the given two dimensional dataset in the fit function
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
```

```python
ytrain.shape
```

```python
LR.fit(xtrain,ytrain)
```

#### Testing Phase:validate the model

```python
xtest=xtest.reshape(-1,1)
ypred=LR.predict(xtest)
```

#### Visualize using pyplot by plotting ypred and ytest versus xtest

```python
plt.scatter(ytest,xtest,color='red',label='ypred')
plt.scatter(ypred,xtest,color='green',label='ytest')
plt.xlabel("Test Input Data")
plt.ylabel("test and pred data")
plt.title("Testing Phase")
plt.legend()
plt.show()
```

#### Parameters of the linear model

```python
LR.coef_
```

```python
LR.intercept_
```

#### Estimate the output of the model for the given input X=30.2

```python
x=30.2
x=np.array(x)
x.shape
x=x.reshape(-1,1)
LR.predict(x)
```

```python
y=LR.coef_*x+LR.intercept_
y
```

#### Performance metrix for the linear model

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(f"Mean Absolute Error of the Linear regression model:{mean_absolute_error(ytest,ypred)}")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,d2_absolute_error_score
a=mean_squared_error(ytest,ypred)
b=mean_absolute_percentage_error(ytest,ypred)
c=d2_absolute_error_score(ytest,ypred)
print(f"Mean Squared Error:{a}")
print(f"Mean Absolute Error:{b}")
print(f"Absolute Error Score:{c}")
```
